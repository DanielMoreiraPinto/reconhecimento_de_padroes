```python
dataset_path = 'C:\\Users\\mvjan\\Desktop\\reconhecimento2\\reconhecimento_de_padroes-feature-noisy-detect\\Mi3_Aligned'
dataset_gaussian_path = 'C:\\Users\\mvjan\\Desktop\\reconhecimento2\\reconhecimento_de_padroes-feature-noisy-detect\\Mi3_Aligned_GaussianNoise'
dataset_sp_path = 'C:\\Users\\mvjan\\Desktop\\reconhecimento2\\reconhecimento_de_padroes-feature-noisy-detect\\Mi3_Aligned_SPNoise'
dataset_poisson_path = 'C:\\Users\\mvjan\\Desktop\\reconhecimento2\\reconhecimento_de_padroes-feature-noisy-detect\\Mi3_Aligned_PoissonNoise'
dataset_speckle_path = 'C:\\Users\\mvjan\\Desktop\\reconhecimento2\\reconhecimento_de_padroes-feature-noisy-detect\\Mi3_Aligned_SpeckleNoise'
```

```python
def gaussian_noise(img, img_name):
    g_noised_image = skimage.util.random_noise(img, mode='gaussian', mean=0, var=1.5)
    #skimage.io.imsave('testeGaussian.jpg',(g_noised_image*255).astype(np.uint8))
    skimage.io.imsave(dataset_gaussian_path + '\\' + img_name, g_noised_image)

def sp_noise(img, img_name):
    sp_noised_image = skimage.util.random_noise(img, mode='s&p', amount=0.5)
    skimage.io.imsave(dataset_sp_path + '\\' + img_name, sp_noised_image)

def poisson_noise(img, img_name):
    poisson_noised_image =random_noise(img, mode='poisson')
    skimage.io.imsave(dataset_poisson_path + '\\' + img_name, poisson_noised_image)
    
def speckle_noise(img, img_name):
    speckle_noised_image =random_noise(img, mode='speckle', mean=0, var=1.5)
    skimage.io.imsave(dataset_speckle_path + '\\' + img_name, speckle_noised_image)
```

```python
i = 0
for pasta_raiz, sub_pastas, arquivos in os.walk(dataset_path):
    for arquivo in arquivos:
        if arquivo.endswith('Reference.bmp'):
            img = skimage.io.imread(fname=os.path.join(pasta_raiz, arquivo))
            gaussian_noise(img, arquivo)
            i = i + 1
print(i)
```

```python
i = 0
for pasta_raiz, sub_pastas, arquivos in os.walk(dataset_path):
    for arquivo in arquivos:
        if arquivo.endswith('Reference.bmp'):
            img = skimage.io.imread(fname=os.path.join(pasta_raiz, arquivo))
            sp_noise(img, arquivo)
            i = i + 1
print(i)
```

```python
i = 0
for pasta_raiz, sub_pastas, arquivos in os.walk(dataset_path):
    for arquivo in arquivos:
        if arquivo.endswith('Reference.bmp'):
            img = skimage.io.imread(fname=os.path.join(pasta_raiz, arquivo))
            poisson_noise(img, arquivo)
            i = i + 1
print(i)
```

```python
i = 0
for pasta_raiz, sub_pastas, arquivos in os.walk(dataset_path):
    for arquivo in arquivos:
        if arquivo.endswith('Reference.bmp'):
            img = skimage.io.imread(fname=os.path.join(pasta_raiz, arquivo))
            speckle_noise(img, arquivo)
            i = i + 1
print(i)
```
